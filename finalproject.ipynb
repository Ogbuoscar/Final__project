Preprocess data, Encode categorical variables

```python
le = LabelEncoder()
mydata['disease_name'] = le.fit_transform(mydata['disease_name'])
mydata['symptom_name'] = le.fit_transform(mydata['symptom_name'])
mydata['sex'] = le.fit_transform(mydata['sex'])
```

Split dataset into train and test sets

```python
features= mydata.drop(columns=['disease_name'])
targets= mydata['disease_name']
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

Train machine learning models/converting to float

```python
mydata['disease_name']=mydata['disease_name'].astype(float)
```

```python
mydata
```

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
svm_model.score(X_test,y_test)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
prediction=dt_model.predict(X_test)
prediction
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
classification_rep=classification_report(y_test,y_pred)
```
